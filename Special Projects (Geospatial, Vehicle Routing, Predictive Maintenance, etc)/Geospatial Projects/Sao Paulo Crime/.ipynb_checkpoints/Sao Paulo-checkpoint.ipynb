```python
import pandas as pd
import numpy as np
import folium
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns 
from shapely.geometry import Point,Polygon,LineString
from folium.plugins import FastMarkerCluster,HeatMap
from folium.features import GeoJson, GeoJsonTooltip, GeoJsonPopup
import branca.colormap as cm

pd.options.display.max_rows = None
pd.options.display.max_columns = None
```

```python
df = pd.read_csv('/Users/danielbeltsazar/DS-ML-DL Mini Projects/Exploring/Geospatial/Sao Paulo Crime/dataset-limpo.csv')
df.head()
```

```python
df.info()
```

```python
df = df.rename(columns={'bairro':'neighborhood','descricao':'description','endereco':'address_of_the_ocurrence',
                       'registrou_bo':'registered_crime','sexo':'sex','tipo_assalto_id':'kind_of_occurence',
                       'titulo':'title','valor_prejuizo':'money_lost','Bicicleta':'bike_stolen',
                       'Bolsa ou Mochila':'backpack_stolen','Carteira':'wallet stolen',
                        'Cartão de Crédito':'creditcard_stolen','Celular':'phone_stolen',
                       'Computador':'computer','Dinheiro':'money','Documentos':'document',
                       'Equipamento de Som':'sound_equipment','Estepe':'spare_car_wheel',
                        'MP4 ou Ipod':'portable_music_player','Móveis':'furniture','Notebook':'laptop',
                       'Outros':'others','Relógio':'wristwatch','Som':'sound'})
```

```python
df['time']=pd.to_datetime(df['time'])
df['created_at']=pd.to_datetime(df['created_at'])
```

```python
geometry = [Point(xy) for xy in zip(df['latitude'],df['longitude'])]
```

```python
geo_df = gpd.GeoDataFrame(df,geometry=geometry)
```

```python
geo_df.crs = "epsg:4326"
geo_df=geo_df.to_crs("EPSG:4326")
```

```python
x_start = (df['latitude'].mean())
y_start = (df['longitude'].mean())
start_coord = (x_start, y_start)

maps = folium.Map(location = start_coord,zoom_start = 8)
geo_df_list = [[point.xy[1][0], point.xy[0][0]] for point in geo_df.geometry]
for coordinates in geo_df_list:
    maps.add_child(folium.Marker(location = coordinates,))
```

```python
my_map = folium.Map(location=[geo_df["latitude"].mean(), geo_df["longitude"].mean()],zoom_start = 5)
my_map.add_child(FastMarkerCluster(geo_df[['latitude', 'longitude']].values.tolist()))
```

```python
map2 = folium.Map(location=[geo_df["latitude"].mean(), geo_df["longitude"].mean()], tiles='cartodbpositron', zoom_start=5)


HeatMap(data=geo_df[['latitude', 'longitude']], radius=10).add_to(map2)


```
